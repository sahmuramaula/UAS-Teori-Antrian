```python
import pandas as pd

# Contoh DataFrame masing-masing lokasi
# Anda ganti ini dengan pembacaan dari CSV Anda
df_st = pd.DataFrame({
    "Hari": ["Sabtu", "Sabtu", "Sabtu", "Minggu", "Minggu", "Minggu", "Senin", "Senin", "Senin"],
    "Periode Waktu": ["18.00-19.00", "19.00-20.00", "20.00-21.00", "18.00-19.00", "19.00-20.00", "20.00-21.00", "18.00-19.00", "19.00-20.00", "20.00-21.00"],
    "Kedatangan": [49, 49, 43, 21, 24, 30, 22, 18, 14]
})

df_al = pd.DataFrame({
    "Hari": ["Sabtu", "Sabtu", "Sabtu", "Minggu", "Minggu", "Minggu", "Senin", "Senin", "Senin"],
    "Periode Waktu": ["18.00-19.00", "19.00-20.00", "20.00-21.00", "18.00-19.00", "19.00-20.00", "20.00-21.00", "18.00-19.00", "19.00-20.00", "20.00-21.00"],
    "Kedatangan": [26, 30, 22, 22, 25, 23, 18, 20, 21]
})

df_n = pd.DataFrame({
    "Hari": ["Sabtu", "Sabtu", "Sabtu", "Minggu", "Minggu", "Minggu", "Senin", "Senin", "Senin"],
    "Periode Waktu": ["18.00-19.00", "19.00-20.00", "20.00-21.00", "18.00-19.00", "19.00-20.00", "20.00-21.00", "18.00-19.00", "19.00-20.00", "20.00-21.00"],
    "Kedatangan": [20, 29, 9, 18, 23, 14, 11, 15, 18]
})

# Gabungkan ketiga DataFrame
df_all = pd.concat([df_st, df_al, df_n])

# Kelompokkan dan jumlahkan berdasarkan Hari dan Periode
df_total = df_all.groupby(["Hari", "Periode Waktu"], as_index=False).sum()

# Atur urutan kategori hari
hari_order = ["Sabtu", "Minggu", "Senin"]
df_total["Hari"] = pd.Categorical(df_total["Hari"], categories=hari_order, ordered=True)

# Urutkan hasilnya jika perlu
df_total = df_total.sort_values(by=["Hari", "Periode Waktu"]).reset_index(drop=True)

print("Jumlah kedatangan total per Hari-Periode:")
print(df_total)
```

```python
import numpy as np

# Hitung rata-rata kedatangan per periode untuk masing-masing stand
mean_st = df_st.groupby("Periode Waktu")["Kedatangan"].mean().rename("Starbucks")
mean_al = df_al.groupby("Periode Waktu")["Kedatangan"].mean().rename("The Allio")
mean_n  = df_n.groupby("Periode Waktu")["Kedatangan"].mean().rename("Nelayan")

# Gabungkan ketiganya ke dalam satu DataFrame
df_mean = pd.concat([mean_st, mean_al, mean_n], axis=1)

# Tambahkan kolom jumlah rata-rata semua stand
df_mean["Jumlah"] = df_mean.sum(axis=1)

# Urutkan index berdasarkan periode waktu agar rapi
periode_order = ["18.00-19.00", "19.00-20.00", "20.00-21.00"]
df_mean = df_mean.reindex(periode_order)

# Hitung jumlah rata-rata untuk masing-masing stand (menurun)
total_st = df_mean["Starbucks"].sum()
total_al = df_mean["The Allio"].sum()
total_n  = df_mean["Nelayan"].sum()
total_jumlah = df_mean["Jumlah"].sum()

# Tambahkan baris ke df_mean
df_mean.loc["Jumlah"] = [total_st, total_al, total_n, np.nan]

# Tampilkan tabel
print("Rata-rata kedatangan pengunjung per periode:")
print(df_mean)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data: ambil rata-rata dari df_mean, exclude baris 'Jumlah'
df_plot = df_mean.drop("Jumlah")

# Siapkan sumbu x
periode = df_plot.index
x = np.arange(len(periode))  # posisi di sumbu x

# Lebar bar
bar_width = 0.25

# Plot setiap stand bergeser di sumbu x
plt.figure(figsize=(10,6))
plt.bar(x - bar_width, df_plot["Starbucks"], width=bar_width, label="Starbucks", color='royalblue')
plt.bar(x, df_plot["The Allio"], width=bar_width, label="The Allio", color='firebrick')
plt.bar(x + bar_width, df_plot["Nelayan"], width=bar_width, label="Nelayan", color='olivedrab')

# Tambahkan nilai angka di atas bar
for i, val in enumerate(df_plot["Starbucks"]):
    plt.text(x[i]-bar_width, val+0.5, f"{val:.1f}", ha='center', fontsize=9)
for i, val in enumerate(df_plot["The Allio"]):
    plt.text(x[i], val+0.5, f"{val:.1f}", ha='center', fontsize=9)
for i, val in enumerate(df_plot["Nelayan"]):
    plt.text(x[i]+bar_width, val+0.5, f"{val:.1f}", ha='center', fontsize=9)

# Tambahkan keterangan
plt.xticks(x, periode)
plt.title("Rata-rata Kedatangan Pengunjung", fontsize=16, fontweight='bold')
plt.ylabel("Rata-rata Kedatangan")
plt.legend(title="Stand")
plt.tight_layout()

plt.show()
```

```python
from IPython.display import display, Math
import math

# Diketahui
lmbda = 70
mu = 40
c = 3

rho = lmbda / (c * mu)
lmbda_mu = lmbda / mu

# Hitung P0
sum_terms = sum([(lmbda_mu**n) / math.factorial(n) for n in range(c)])
last_term = (lmbda_mu**c) / (math.factorial(c) * (1 - rho))
P0 = 1 / (sum_terms + last_term)

display(Math(r'P_0 \;(\text{probabilitas sistem kosong}) = \frac{1}{\sum_{n=0}^{c-1}\frac{\lambda}{\mu}^n / n! + \frac{\lambda}{\mu}^c / (c! \times (1 - \rho))} = %.4f' % P0))
```

```python
Lq = ( (lmbda_mu)**c * rho ) / ( math.factorial(c) * (1 - rho)**2 ) * P0

display(Math(r'L_q \;(\text{rata-rata pelanggan dalam antrian}) = \frac{\frac{\lambda}{\mu}^c \times \rho}{c! \times (1 - \rho)^2} \times P_0 = %.2f' % Lq))
```

```python
Ls = Lq + lmbda / mu

display(Math(r'L_s \;(\text{rata-rata pelanggan dalam sistem}) = L_q + \frac{\lambda}{\mu} = %.2f' % Ls))
```

```python
Wq = Lq / lmbda
Wq_min = Wq * 60

display(Math(r'W_q \;(\text{waktu tunggu rata-rata di antrian}) = \frac{L_q}{\lambda} = %.2f \;\text{jam} = %.2f \;\text{menit}' % (Wq, Wq_min)))
```

```python
Ws = Wq + 1 / mu
Ws_min = Ws * 60

display(Math(r'W_s \;(\text{waktu rata-rata dalam sistem}) = W_q + \frac{1}{\mu} = %.2f \;\text{jam} = %.2f \;\text{menit}' % (Ws, Ws_min)))
```

```python
Pc = (lmbda_mu**c) / (math.factorial(c) * (1 - rho)) * P0

display(Math(r'P_c \;(\text{probabilitas semua server sibuk}) = \frac{\frac{\lambda}{\mu}^c}{c! \times (1 - \rho)} \times P_0 = %.4f' % Pc))
```
